```python
import pandas as pd
import numpy as np
import datetime
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
cat_df = pd.read_csv('../lab-cleaning-categorical-data/files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
cat_df.head()
```

```python
cat_df.dtypes
```

```python
cat_df.columns = cat_df.columns.str.lower()
cat_df.columns = cat_df.columns.str.replace(' ', '_')
cat = cat_df.select_dtypes(include = np.object)
cat
```

```python
cat.info()
```

```python
cat = cat.set_index('customer')
```

```python
#cat2 = cat.copy()
#cat2=pd.get_dummies(cat, drop_first=True)
#cat2
```

```python
for col in cat.columns:
    sns.barplot(x=cat[col].value_counts().index, y=cat[col].value_counts(), data=cat)
    plt.show()
```

```python
for col in cat.columns:
    sns.displot(x=cat[col])#.value_counts().index, y=cat[col].value_counts(), data=cat)
    plt.show()
```

```python
#I do not see any redundant data
for column in cat.columns:
    print(column, cat[column].unique())
```

```python
#creating new columns month, day, year
cat['month'] = cat['effective_to_date'].str.split('/').str[0]
cat['day'] = cat['effective_to_date'].str.split('/').str[1]
cat['year'] = cat['effective_to_date'].str.split('/').str[2]
cat
```

```python
#converting to datetime format
cat['effective_to_date'] = cat['effective_to_date'].apply(pd.to_datetime)
```

```python
cat['month'] = pd.DatetimeIndex(cat['effective_to_date']).month
```
